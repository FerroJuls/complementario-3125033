```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Ruta del archivo CSV (actualiza la ruta si lo necesitas)
archivo_csv = "./drive/MyDrive/csv/8.HerramientasDeBúsquedaYDesarrolloDeIA/8.HerramientasDeBúsquedaYDesarrolloDeIA.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Limpiar los datos (si es necesario)
df.columns = df.columns.str.strip()
df['Herramienta de Búsqueda e IA'] = df['Herramienta de Búsqueda e IA'].str.strip()

# Eliminar el símbolo de porcentaje y convertir a flotante
df['Porcentaje (%)'] = df['Porcentaje (%)'].replace({'%': ''}, regex=True).astype(float)

# Ordenar por el Porcentaje (%)
df = df.sort_values(by='Porcentaje (%)', ascending=False)

# Configuración del gráfico
plt.figure(figsize=(14, 8))
bar_width = 0.4
index = range(len(df))

plt.bar(index, df['Porcentaje (%)'], bar_width, label='Porcentaje (%)', color='skyblue')

X = np.array(range(len(df))).reshape(-1, 1)  # Indices de las tecnologías
y = df['Porcentaje (%)'].values  # Porcentajes

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.plot(index, y_pred, color='red', label='Regresión Lineal', linewidth=2)

plt.xlabel('Herramienta de Búsqueda e IA', fontsize=14)
plt.ylabel('Porcentaje (%)', fontsize=14)
plt.title('Comparación de Popularidad de Herramienta de Búsqueda e IA con Regresión Lineal', fontsize=16)
plt.xticks([i for i in index], df['Herramienta de Búsqueda e IA'], rotation=90)
plt.legend()

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()
```
