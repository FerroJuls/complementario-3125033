```python
import matplotlib.pyplot as plt
import pandas as pd

# Ruta del archivo CSV
archivo_csv = f"./drive/MyDrive/csv/1.LenguagesPopularesTIBOE/TodosLosAños.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Asegurarse de que las columnas están correctamente formateadas
df.columns = df.columns.str.strip().str.capitalize()

# Limpiar y convertir los valores de los años
años = [str(año) for año in range(2001, 2025)]  # Lista de años en las columnas
for año in años:
    df[año] = df[año].str.rstrip('%').astype(float)  # Eliminar el símbolo '%' y convertir a float

# Reorganizar los datos para la gráfica
df_melted = df.melt(id_vars=['Lenguaje'], var_name='Año', value_name='Popularidad (%)')

# Asegurarse de que los años estén en orden numérico
df_melted['Año'] = df_melted['Año'].astype(int)

# Generar el gráfico de línea para cada lenguaje
plt.figure(figsize=(14, 8))
for lenguaje in df['Lenguaje'].unique():
    subset = df_melted[df_melted['Lenguaje'] == lenguaje]
    plt.plot(subset['Año'], subset['Popularidad (%)'], marker='o', label=lenguaje)

# Configurar el gráfico
plt.title('Popularidad de lenguajes de programación según TIOBE (2001-2024)', fontsize=16)
plt.xlabel('Años', fontsize=14)
plt.ylabel('Popularidad (%)', fontsize=14)
plt.xticks(range(2001, 2025), rotation=45)
plt.legend(title='Lenguajes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Aca coloca e año a investigar del 2001 al 2024 este lo tomara en la url y la grafica cambiar segun lo necesitado
año = 2004
archivo_csv = f"./drive/MyDrive/csv/1.LenguagesPopularesTIBOE/{año}.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Asegurar que las columnas de meses están correctamente formateadas
df.columns = df.columns.str.strip().str.capitalize()

# Limpiar y convertir los valores de las columnas de meses
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
for mes in meses:
    df[mes] = df[mes].str.replace(r"[^\d\.%]", "", regex=True)  # Eliminar caracteres no válidos
    df[mes] = df[mes].str.rstrip('%').astype(float)  # Eliminar el símbolo '%' y convertir a float

# Reorganizar los datos para la gráfica
df_melted = df.melt(id_vars=['Año', 'Lenguaje'], var_name='Mes', value_name='Popularidad (%)')

# Ordenar los meses cronológicamente
meses_orden = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
df_melted['Mes'] = pd.Categorical(df_melted['Mes'], categories=meses_orden, ordered=True)

# Generar el gráfico de dispersión con líneas
plt.figure(figsize=(14, 8))
for lenguaje in df['Lenguaje'].unique():
    subset = df_melted[df_melted['Lenguaje'] == lenguaje]
    plt.plot(subset['Mes'], subset['Popularidad (%)'], marker='o', label=lenguaje)

# Configurar el gráfico
plt.title('Popularidad de lenguajes de programación según TIOBE', fontsize=16)
plt.xlabel(f'Mes del Año {año}', fontsize=14)
plt.ylabel('Popularidad (%)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Lenguajes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
