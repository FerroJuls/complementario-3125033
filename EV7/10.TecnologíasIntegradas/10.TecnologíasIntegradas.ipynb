```python
import matplotlib.pyplot as plt
import pandas as pd

# Ruta del archivo CSV (actualiza la ruta si lo necesitas)
archivo_csv = "./drive/MyDrive/csv/10.TecnologíasIntegradas/10.TecnologíasIntegradas.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Limpiar los datos (si es necesario)
df.columns = df.columns.str.strip()
df['Tecnología Embebida'] = df['Tecnología Embebida'].str.strip()

# Ordenar por el porcentaje deseado
df = df.sort_values(by='Deseada (%)', ascending=False)

# Configuración del gráfico
plt.figure(figsize=(14, 8))
bar_width = 0.4
index = range(len(df))

# Barras para "Deseado" y "Admirado"
plt.bar(index, df['Deseada (%)'], bar_width, label='Deseada (%)', color='skyblue')
plt.bar([i + bar_width for i in index], df['Admirada (%)'], bar_width, label='Admirada (%)', color='orange')

# Configuración de etiquetas
plt.xlabel('Tecnología Embebida', fontsize=14)
plt.ylabel('Porcentaje (%)', fontsize=14)
plt.title('Comparación de popularidad: Deseado vs Admirado', fontsize=16)
plt.xticks([i + bar_width / 2 for i in index], df['Tecnología Embebida'], rotation=90)
plt.legend()

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()
```
