```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Ruta del archivo CSV (actualiza la ruta si lo necesitas)
archivo_csv = "./drive/MyDrive/csv/4.MarcosTecnologiasWeb/4.MarcosTecnologiasWeb.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Limpiar los datos (si es necesario)
df.columns = df.columns.str.strip()
df['Tecnología'] = df['Tecnología'].str.strip()

# Eliminar el símbolo de porcentaje y convertir a flotante
df['Porcentaje'] = df['Porcentaje'].replace({'%': ''}, regex=True).astype(float)

# Ordenar por el porcentaje
df = df.sort_values(by='Porcentaje', ascending=False)

# Configuración del gráfico
plt.figure(figsize=(14, 8))
bar_width = 0.4
index = range(len(df))

# Barras para el porcentaje de cada tecnología
plt.bar(index, df['Porcentaje'], bar_width, label='Porcentaje', color='skyblue')

# Preparar los datos para la regresión lineal
X = np.array(range(len(df))).reshape(-1, 1)  # Indices de las tecnologías
y = df['Porcentaje'].values  # Porcentajes

# Crear el modelo de regresión lineal y ajustarlo
model = LinearRegression()
model.fit(X, y)

# Predecir valores con el modelo de regresión
y_pred = model.predict(X)

# Dibujar la línea de regresión
plt.plot(index, y_pred, color='red', label='Regresión Lineal', linewidth=2)

# Configuración de etiquetas
plt.xlabel('Tecnologías', fontsize=14)
plt.ylabel('Porcentaje (%)', fontsize=14)
plt.title('Comparación de Popularidad de Tecnologías con Regresión Lineal', fontsize=16)
plt.xticks([i for i in index], df['Tecnología'], rotation=90)
plt.legend()

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()

```
