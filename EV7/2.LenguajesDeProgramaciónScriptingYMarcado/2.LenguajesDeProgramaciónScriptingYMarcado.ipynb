```python
import matplotlib.pyplot as plt
import pandas as pd

# Ruta del archivo CSV (actualiza la ruta si lo necesitas)
archivo_csv = "./drive/MyDrive/csv/2.LenguajesDeProgramaciónScriptingYMarcado/2.LenguajesDeProgramaciónScriptingYMarcado.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Limpiar los datos (si es necesario)
df.columns = df.columns.str.strip()
df['Lenguaje'] = df['Lenguaje'].str.strip()

# Ordenar por el porcentaje deseado
df = df.sort_values(by='Deseado (%)', ascending=False)

# Configuración del gráfico
plt.figure(figsize=(14, 8))
bar_width = 0.4
index = range(len(df))

# Barras para "Deseado" y "Admirado"
plt.bar(index, df['Deseado (%)'], bar_width, label='Deseado (%)', color='skyblue')
plt.bar([i + bar_width for i in index], df['Admirado (%)'], bar_width, label='Admirado (%)', color='orange')

# Configuración de etiquetas
plt.xlabel('Lenguajes de programación', fontsize=14)
plt.ylabel('Porcentaje (%)', fontsize=14)
plt.title('Comparación de popularidad: Deseado vs Admirado', fontsize=16)
plt.xticks([i + bar_width / 2 for i in index], df['Lenguaje'], rotation=90)
plt.legend()

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()
```
