```python
import matplotlib.pyplot as plt
import pandas as pd

# Aca coloca e ao a investigar del 2001 al 2024 este lo tomara en la url y la grafica cambiar segun lo necesitado
ao = 2010
archivo_csv = f"./drive/MyDrive/csv/1.LenguagesPopularesTIBOE/{ao}.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Asegurar que las columnas de meses estn correctamente formateadas
df.columns = df.columns.str.strip().str.capitalize()

# Limpiar y convertir los valores de las columnas de meses
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
for mes in meses:
    df[mes] = df[mes].str.replace(r"[^\d\.%]", "", regex=True)  # Eliminar caracteres no vlidos
    df[mes] = df[mes].str.rstrip('%').astype(float)  # Eliminar el smbolo '%' y convertir a float

# Reorganizar los datos para la grfica
df_melted = df.melt(id_vars=['Ao', 'Lenguaje'], var_name='Mes', value_name='Popularidad (%)')

# Ordenar los meses cronolgicamente
meses_orden = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
df_melted['Mes'] = pd.Categorical(df_melted['Mes'], categories=meses_orden, ordered=True)

# Generar el grfico de dispersin con lneas
plt.figure(figsize=(14, 8))
for lenguaje in df['Lenguaje'].unique():
    subset = df_melted[df_melted['Lenguaje'] == lenguaje]
    plt.plot(subset['Mes'], subset['Popularidad (%)'], marker='o', label=lenguaje)

# Configurar el grfico
plt.title('Popularidad de lenguajes de programacin segn TIOBE', fontsize=16)
plt.xlabel(f'Mes del Ao {ao}', fontsize=14)
plt.ylabel('Popularidad (%)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Lenguajes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ruta del archivo CSV
archivo_csv = f"./drive/MyDrive/csv/1.LenguagesPopularesTIBOE/TodosLosAos.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Asegurarse de que las columnas estn correctamente formateadas
df.columns = df.columns.str.strip().str.capitalize()

# Limpiar y convertir los valores de los aos
aos = [str(ao) for ao in range(2001, 2025)]  # Lista de aos en las columnas
for ao in aos:
    df[ao] = df[ao].str.rstrip('%').astype(float)  # Eliminar el smbolo '%' y convertir a float

# Reorganizar los datos para la grfica
df_melted = df.melt(id_vars=['Lenguaje'], var_name='Ao', value_name='Popularidad (%)')

# Asegurarse de que los aos estn en orden numrico
df_melted['Ao'] = df_melted['Ao'].astype(int)

# Generar el grfico de lnea para cada lenguaje
plt.figure(figsize=(14, 8))
for lenguaje in df['Lenguaje'].unique():
    subset = df_melted[df_melted['Lenguaje'] == lenguaje]
    plt.plot(subset['Ao'], subset['Popularidad (%)'], marker='o', label=lenguaje)

# Configurar el grfico
plt.title('Popularidad de lenguajes de programacin segn TIOBE (2001-2024)', fontsize=16)
plt.xlabel('Aos', fontsize=14)
plt.ylabel('Popularidad (%)', fontsize=14)
plt.xticks(range(2001, 2025), rotation=45)
plt.legend(title='Lenguajes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Mostrar el grfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Ruta del archivo CSV (actualiza la ruta si lo necesitas)
archivo_csv = "./drive/MyDrive/csv/4.MarcosTecnologiasWeb/4.MarcosTecnologiasWeb.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Limpiar los datos (si es necesario)
df.columns = df.columns.str.strip()
df['Tecnologa'] = df['Tecnologa'].str.strip()

# Eliminar el smbolo de porcentaje y convertir a flotante
df['Porcentaje'] = df['Porcentaje'].replace({'%': ''}, regex=True).astype(float)

# Ordenar por el porcentaje
df = df.sort_values(by='Porcentaje', ascending=False)

# Configuracin del grfico
plt.figure(figsize=(14, 8))
bar_width = 0.4
index = range(len(df))

# Barras para el porcentaje de cada tecnologa
plt.bar(index, df['Porcentaje'], bar_width, label='Porcentaje', color='skyblue')

# Preparar los datos para la regresin lineal
X = np.array(range(len(df))).reshape(-1, 1)  # Indices de las tecnologas
y = df['Porcentaje'].values  # Porcentajes

# Crear el modelo de regresin lineal y ajustarlo
model = LinearRegression()
model.fit(X, y)

# Predecir valores con el modelo de regresin
y_pred = model.predict(X)

# Dibujar la lnea de regresin
plt.plot(index, y_pred, color='red', label='Regresin Lineal', linewidth=2)

# Configuracin de etiquetas
plt.xlabel('Tecnologas', fontsize=14)
plt.ylabel('Porcentaje (%)', fontsize=14)
plt.title('Comparacin de Popularidad de Tecnologas con Regresin Lineal', fontsize=16)
plt.xticks([i for i in index], df['Tecnologa'], rotation=90)
plt.legend()

# Ajustar diseo y mostrar
plt.tight_layout()
plt.show()
```
