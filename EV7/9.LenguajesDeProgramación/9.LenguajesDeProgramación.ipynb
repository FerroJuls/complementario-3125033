```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Ruta del archivo CSV (actualiza la ruta si lo necesitas)
archivo_csv = "./drive/MyDrive/csv/9.LenguajesDeProgramación/9.LenguajesDeProgramación.csv"

# Leer el archivo CSV
df = pd.read_csv(archivo_csv)

# Limpiar los datos (si es necesario)
df.columns = df.columns.str.strip()
df['Programming Language'] = df['Programming Language'].str.strip()

# Eliminar el símbolo de Popularity y convertir a flotante
df['Popularity (%)'] = df['Popularity (%)'].replace({'%': ''}, regex=True).astype(float)

# Ordenar por el Popularity (%)
df = df.sort_values(by='Popularity (%)', ascending=False)

# Configuración del gráfico
plt.figure(figsize=(14, 8))
bar_width = 0.4
index = range(len(df))

plt.bar(index, df['Popularity (%)'], bar_width, label='Popularity (%)', color='skyblue')

X = np.array(range(len(df))).reshape(-1, 1)  # Indices de las tecnologías
y = df['Popularity (%)'].values  # Popularitys

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.plot(index, y_pred, color='red', label='Regresión Lineal', linewidth=2)

plt.xlabel('Programming Language', fontsize=14)
plt.ylabel('Popularity (%)', fontsize=14)
plt.title('Comparación de Popularidad de  Programming Language con Regresión Lineal', fontsize=16)
plt.xticks([i for i in index], df['Programming Language'], rotation=90)
plt.legend()

# Ajustar diseño y mostrar
plt.tight_layout()
plt.show()
```
